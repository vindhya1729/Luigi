#Step 1: Import the Iris Dataset, Part-2

```python

mydata = pd.read_csv("/content/drive/MyDrive/bigdatahw4/iris.data", header=None)

mydata.columns =['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)','class']
mydata
```

```python
my_iris = spark.createDataFrame(mydata)

my_spark_iris = my_iris.drop("class")
my_spark_iris.show()
```

#Step 2: Building a K-means model to classify the species of Iris.
You can choose a k value randomly at this step.

```python
from pyspark.ml.feature import VectorAssembler
assemble=VectorAssembler(inputCols=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)'],
                         outputCol = 'iris_features')
assembled_data=assemble.transform(my_spark_iris)
assembled_data.show()
```

```python
from pyspark.ml.clustering import KMeans
from pyspark.ml.evaluation import ClusteringEvaluator

KMeans1=KMeans(featuresCol='iris_features', k=3)
KMeans_Model=KMeans1.fit(assembled_data)

KMeans_Assignments=KMeans_Model.transform(assembled_data)
KMeans_Assignments.show()
```

#Step 3: Use the Silhouette score to report the original performance.
The silhouette score is a statistic for evaluating clustering techniques. It's a metric for how similar a data point is to the other points in a cluster. The better the grouping, the higher the silhouette score. The number of clusters (k) is the hyper-parameter to be tuned in the k means clustering algorithm.

```python
org_evaluator = ClusteringEvaluator(featuresCol='iris_features',metricName='silhouette', distanceMeasure='squaredEuclidean')

my_silhouette = org_evaluator.evaluate(KMeans_Assignments)
print("Silhouette Score = " + str(my_silhouette))
```

#Step 4: Try to improve the performance of the original model by trying at least 10 different k values

```python
silhouette_max = 0
kmax = 0
evaluator = ClusteringEvaluator(featuresCol='iris_features', metricName='silhouette', distanceMeasure='squaredEuclidean')

for K in range(2,13):

    kmeans=KMeans(featuresCol='iris_features', k=K)
    model=kmeans.fit(assembled_data)
    KMeans_predictions=model.transform(assembled_data)


    silhouette_score=evaluator.evaluate(KMeans_predictions)
    print("k = "+ str(K) + ",Silhouette Score = " + str(silhouette_score))

    if silhouette_score > silhouette_max:
      silhouette_max = silhouette_score
      kmax = K
```

#Step 5: Select the best 'k' from the above step and print the following code:

```python
print("k="+ str(kmax) + " gives the best performance, Silhouette = "+ str(silhouette_max))
```
